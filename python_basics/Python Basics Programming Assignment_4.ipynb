```python
def fact(n):
    if n==0 or n==1:
        print ("factorial of",n,"is 1")
    else:
        f=1
        for i in range(1,n+1):
            f=f*i
        print("factorial of",n,"is",f)
        
fact(n= int(input("enter a number: ")))
```

```python
def table(n):
    if n==0:
        print("the output will remain 0 if multiplied by 0")
    else:
        for i in range(1,11):
            a=n*i
            print (n, "x", i, "=" ,a)
        
table(n= int(input("enter the number for which you want multiplication table: ")))
```

```python
def fibonacci(n):   
    a=0
    b=1
    print(a,b,end=" ")
    for i in range(n):
        c=a+b
        a=b
        b=c
        print (c, end=" ")
        
fibonacci(n=int(input("Enter the series you want: ")))
```

```python
def armstrong(n):
    if n<=0:
        print(n,"is not an armstrong number")
    else:
        temp= n
        l= len(str(n))
        sum1=0
        while temp!=0:
            k= temp%10
            sum1 += k**l
            temp= temp//10
        if sum1==n:
            print(n, "is an armstrong number")
        else:
            print(n, "is not an armstrong number")

armstrong(n= int(input("Enter a number to check if it's armstrong or not: ")))
```

```python
def armstrong(s,e):
    
    for i in range(s,e):
        temp= i
        l= len(str(i))
        sum1= 0
        while temp!=0:
            k= temp%10
            sum1 += k**l
            temp= temp//10
        if i==sum1:
            print(i,"is an armstrong number")
        else:
            print(i,"is not an armstrong number")
            
s= int(input("Enter starting range: "))
if s<=0:
    print(s,"is not an armstrong number")
else:
    e= int(input("Enter ending range: "))
    armstrong(s,e)
```
